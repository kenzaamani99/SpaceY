# Web Scraping

Le Web Scraping (grattage) est une technique utilisée par les développeurs pour extraire  (littéralement « gratter ») des informations d’un site Internet.  

# How Does Web Scraping Work?
1. <b>Request-Response: </b>
    la Lib requests permet de récupérer les pages Web via le protocole http(s)
2. <b>Parse and Extract: </b>
    la Lib BeautifulSoup va analyser ces pages et extraire les données 
3. <b>Manipulate and save Data: </b> la Lib pandas pour la manipulation et le stockage de ces données.

```python
# requests: make HTTP requests
# Beautiful: library for parsing HTML and XML documents; in order to extract data.
# Pandas: library for data manipulation.
import  requests
from  bs4 import BeautifulSoup
import  pandas as pd
page=requests.get('https://forecast.weather.gov/MapClick.php?x=190&y=120&site=okx&zmx=&zmy=&map_x=189&map_y=119#.XWpFOJNKg1I')
soup=BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
week = soup.find(id='seven-day-forecast-container')
print(week.prettify())
```

```python
items=week.find_all(class_='tombstone-container')
#print(items[0])
period_names=[item.find(class_='period-name').get_text() for item in items]
short_descriptions=[item.find(class_='short-desc').get_text() for item in items]
temperatures=[item.find(class_='temp').get_text() for item in items]

print(period_names)
print(short_descriptions)
print(temperatures)
```

```python
weather_stuff=pd.DataFrame(
    {
        'period':period_names,
        'short_descriptions':short_descriptions,
        'temperatures': temperatures,
    }
)
print(weather_stuff)
```

```python
weather_stuff.to_csv('weather_m11_22.csv')
```

```python
weather_stuff.to_json('weather_m11_22.json')
```

```python
import requests 
from bs4 import BeautifulSoup 
import pandas as pd  
names=[] 
prices=[] 

for i in range(1,11):         
    page = requests.get("https://www.jumia.ma/smartphones-android/?page="+str(i))      
    soup = BeautifulSoup(page.content, 'html.parser')      
    items = soup.find_all('div', attrs={'class': 'info'})      
    names=names + [item.find(class_='name').get_text() for item in items]     
    prices=prices + [item.find(class_='prc').get_text() for item in items] 
    print('Web scarping ......... page:',i )
df=pd.DataFrame({
    'nom de telephone':names,
    'prix':prices,
} )  
df.to_csv('telephones.csv')
```
